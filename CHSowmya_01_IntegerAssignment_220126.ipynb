```python
# Question 5: Calculate the sum of digits in number 12345
print("\nQuestion 5: Calculate the sum of digits in number 12345")

# Way 1: Using string conversion + loop
number = 12345
total1 = 0
for digit in str(number):
    total1 += int(digit)
print("Way 1:", total1)

# Way 2: Using sum() + map()
number = 12345
total2 = sum(map(int, str(number)))
print("Way 2:", total2)

# Way 3: Using mathematical approach
number = 12345
total3 = 0
while number > 0:
    total3 += number % 10
    number //= 10
print("Way 3:", total3)
```

```python
# Question 6: Check if 97 is a prime number
print("\nQuestion 6: Check if 97 is a prime number")
# Your code here
num = 97
if num <=1:
  print(f'{num} is not a prime number')
else:
  for i in range(2, int(num ** 0.5) + 1):
    if num % i == 0:
      print(f'{num} is not a prime number')
      break

print(f'{num} is a prime number')
```

```python
# Question 9: Find the greatest common divisor (GCD) of 48 and 36
print("\nQuestion 9: Find the greatest common divisor (GCD) of 48 and 36")
# Your code here
a = 48
b = 36

while b != 0:
    a, b = b, a % b

print(a)
```
