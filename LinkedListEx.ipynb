#Linked List

What is a Linked List?
- A Linked List is a data structure used for storing collections of data.
- Each element or node of a list is comprising of two items:
  - Data
  - Pointer(reference) to the next node
- In a Linked List, the elements are not stored at contiguous (in simple words, continuous) memory locations.
- The first node of a linked list is known as "Head".
- The last node of a linked list is known as "Tail".
- The last node has a reference to "None".

#1. Reverse Linked List (LeetCode 206)
**Question**

You are given the head of a singly linked list. Reverse the list and return the new head.

**DSA Pattern**

Pointer Manipulation (Iterative Reversal)

**Approach**

Traverse the list once and reverse the direction of each node’s next pointer.

**Pseudocode**


image.png


```python
def reverseList(head):
    prev = None
    curr = head

    while curr:
        nxt = curr.next
        curr.next = prev
        prev = curr
        curr = nxt

    return prev
```

```python
# DETAILED SOLUTION
def reverseList(head):
  #head is the first element of the list so we are taking that only
  # Here the two pointers are prev and curr
  # prev - Holds the prev data of the linkedList
  # curr - Holds the current data of the linkedList
  # As we are starting from start of the list there is no prev and current is head

  #First Pointer
  prev = None
  #Second pointer
  curr = head

  # the loop transveres(travels) the whole linked list
  while curr:
    nxt = curr.next # Saving current node address
    curr.next = prev # Assigning the current node address to previous one so that it can be reversed
    prev = curr # Update prev
    curr = nxt # Update current

  return prev
# ------------------------------------------------------------------------------------------------------
#Defining Node
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

# Giving input manually
# Create nodes
n1 = Node(1)
n2 = Node(2)
n3 = Node(3)

# Link nodes
n1.next = n2
n2.next = n3

head = n1   # first node

# Reverse
head = reverseList(head)

#Print LinkedList
def printList(head):
    temp = head
    while temp:
        print(temp.data, end=" -> ")
        temp = temp.next
    print("None")

printList(head)
```

# 2.  Merge Two Sorted Lists (LeetCode 21)
**Question**

Merge two sorted linked lists into one sorted list.

Input:
l1 = 1 → 3 → 5
l2 = 2 → 4 → 6

Output:
1 → 2 → 3 → 4 → 5 → 6

**DSA Pattern**

Dummy Node + Two Pointers

**Approach**

- Use a dummy node and compare nodes from both lists.
- Dummy avoids special handling for head
- Always advance the pointer you used

**Pseudocode**

image.png

```python
def mergeTwoLists(l1, l2):
    dummy = ListNode(0)
    tail = dummy

    while l1 and l2:
        if l1.val <= l2.val:
            tail.next = l1
            l1 = l1.next
        else:
            tail.next = l2
            l2 = l2.next
        tail = tail.next

    tail.next = l1 if l1 else l2
    return dummy.next
```

```python
#DETAILED SOLUTION
def mergeTwoLists(l1,l2):
  dummy = ListNode(0) #Fake starting node. Not a part of actual merged list.
  '''
  Why do we need it?

  Because when building a new linked list, the first node is always tricky.
  Without dummy, you’d need logic like:
  “If head is None…”
  “Else attach…”

  dummy avoids all that.

  Think of it as:
  🧷 A temporary hook to hang the list while building it
  At the end, we simply skip it by writing following code
  return dummy.next

  '''
  tail = dummy
  '''
  tail is the pointer that always points to the last node of the merged list so far

  tail.next = something → attach a node
  tail = tail.next → move forward

  In simple words:
  tail is the hand that keeps extending the list
  '''

  while l1 and l2: #Loop runs till both the list have values in it
    if l1.val <= l2.val:
      #l1 value is smaller than l2 value
      tail.next = l1 #update tail so that you can move to next node
      l1 = l1.next #going to next node in list 1

    else:
      #l2 value is smaller than l1 value
      tail.next = l2
      l2 = l2.next #going to next node in list 2

    tail = tail.next

  #What if list 1 has more values than list 2? Or vice-versa? We just have to add them at the end of the mergedList
  tail.next = l1 if l1 else l2
  '''
  The above line can be interpreted as
  if l1:
    tail.next = l1
  else:
    tail.next = l2
  '''

  return dummy.next #dummy.next has the first element of the merged list

#-----------------------------------------------------
#Defining Node
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
# List 1: 1 -> 3 -> 5
a1 = ListNode(1)
a2 = ListNode(3)
a3 = ListNode(5)
a1.next = a2
a2.next = a3
l1 = a1

# List 2: 2 -> 4 -> 6
b1 = ListNode(2)
b2 = ListNode(4)
b3 = ListNode(6)
b1.next = b2
b2.next = b3
l2 = b1

# Merge
head = mergeTwoLists(l1, l2)

def printList(head):
    temp = head
    while temp:
        print(temp.val, end=" -> ")
        temp = temp.next
    print("None")

printList(head)

```

#3. Remove Nth Node From End (LeetCode 19)

**Question**

Given a linked list, remove the Nth node from the end and return the updated list.

- **🔍 Key clarifications**

  - You don’t know the length of the list beforehand
  - You must do it in one pass
  - n is always valid
- **Example**

Input : 1 → 2 → 3 → 4 → 5, n = 2

Output: 1 → 2 → 3 → 5

**DSA Pattern**

Two Pointers + Dummy Node

**Approach**

Maintain a gap of n between two pointers.

**Pseudocode**

dummy → head

fast = slow = dummy

move fast n steps

move both till fast reaches end

remove slow.next


```python
def removeNthFromEnd(head, n):
    dummy = ListNode(0, head)
    fast = slow = dummy

    for _ in range(n):
        fast = fast.next

    while fast.next:
        fast = fast.next
        slow = slow.next

    slow.next = slow.next.next
    return dummy.next
```

```python
def removeNthFromEnd(head,n) :
  #Create a fake node before real head.
  #BUT WHY?
  # If we remove the first node, handling becomes tricky
  # Dummy simplifies everything
  # dummy → 1 → 2 → 3 → 4 → 5
  dummy = ListNode(0,head)

  # Why two pointers? To maintain a gap of n nodes between them
  fast = slow = dummy

  #fast is n nodes ahead of slow
  for _ in range(n):
    fast = fast.next

  # Move both pointers until fast reaches last node
  # Why fast.next and not fast?
  # Because we want slow to stop just before the node to delete
  while fast.next:
    # They move together, keeping the distance = n
    fast = fast.next
    slow = slow.next
  # This skips the Nth node from the end
  slow.next = slow.next.next
  # Because dummy is fake
  # The real head starts from dummy.next
  return dummy.next
#-----------------------------------------------------
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
# Create nodes
a = ListNode(1)
b = ListNode(2)
c = ListNode(3)
d = ListNode(4)
e = ListNode(5)

# Link nodes
a.next = b
b.next = c
c.next = d
d.next = e

head = a
n = 2

# Call function
head = removeNthFromEnd(head, n)

def printList(head):
    while head:
        print(head.val, end=" -> ")
        head = head.next
    print("None")

printList(head)

```

# 4. Linked List Cycle (LeetCode 141)
Determine whether a linked list contains a cycle.

A cycle exists if:

A node’s next points to a previous node, forming a loop

**🔍 Key clarifications**

You must not modify the list

Return only True or False

You don’t need to find where the cycle starts

**📌 Example**

1 → 2 → 3 → 4

      ↑     ↓

      ← ← ←
Output: True

**DSA Pattern**

Floyd’s Cycle Detection (Fast & Slow Pointers)

**Approach**

Fast pointer moves twice as fast as slow.

**Pseudocode**

image.png

```python
def hasCycle(head):
    slow = fast = head

    while fast and fast.next:
        slow = slow.next
        fast = fast.next.next
        if slow == fast:
            return True

    return False
```

```python
#Detailed Solution
'''
What is a cycle?

A cycle means:
While moving forward in the list,you never reach None
because one node points back to a previous node.

Core idea (Floyd’s Cycle Detection / Tortoise & Hare)

slow pointer → moves 1 step at a time
fast pointer → moves 2 steps at a time

👉 If there is a cycle, fast will eventually meet slow
👉 If there is NO cycle, fast will reach None
'''
def hasCycle(head):
    # Step 1: Initialize two pointers at the head of the list
    slow = fast = head

    # Step 2: Loop runs as long as fast pointer can move forward safely
    while fast and fast.next:

        # slow moves one step
        slow = slow.next

        # fast moves two steps
        fast = fast.next.next

        # Step 3: If slow and fast meet, a cycle exists
        if slow == fast:
            return True

    # Step 4: If fast reaches None, no cycle exists
    return False

#--------------------------------------------------------------------
class ListNode:
    def __init__(self, val):
        self.val = val
        self.next = None
# WITHOUT CYCLE
# Create nodes
head = ListNode(1)
second = ListNode(2)
third = ListNode(3)
fourth = ListNode(4)

# Link nodes
head.next = second
second.next = third
third.next = fourth
fourth.next = None   # explicitly showing end

# Test
print(hasCycle(head))  # Output: False

# WITH CYCLE

# Create nodes
head1 = ListNode(1)
second1 = ListNode(2)
third1 = ListNode(3)
fourth1 = ListNode(4)

# Link nodes
head1.next = second1
second1.next = third1
third1.next = fourth1

# Create a cycle here
fourth1.next = second1   # cycle formed

# Test
print(hasCycle(head1))  # Output: True

```

# 5. Add Two Numbers (LeetCode 2)
❓ **What is the question asking?**

Two linked lists represent two non-negative integers.
Each node stores one digit, and digits are stored in reverse order.

Add the two numbers and return the sum as a linked list.

🔍 **Key clarifications**

Digits are reversed

You must handle carry

Result should also be in reverse order

📌 **Example**
Input:
l1 = 2 → 4 → 3  (represents 342)

l2 = 5 → 6 → 4  (represents 465)

**Output:**

7 → 0 → 8       (represents 807)

**DSA Pattern**

Dummy Node + Carry Handling

**Approach**

Simulate digit-by-digit addition like math.

**Pseudocode**

image.png

```python
def addTwoNumbers(l1, l2):
    dummy = ListNode(0)
    curr = dummy
    carry = 0

    while l1 or l2 or carry:
        v1 = l1.val if l1 else 0
        v2 = l2.val if l2 else 0

        total = v1 + v2 + carry
        carry = total // 10

        curr.next = ListNode(total % 10)
        curr = curr.next

        if l1: l1 = l1.next
        if l2: l2 = l2.next

    return dummy.next
```

```python
''''
Beginner Mental Model

Think like adding numbers on paper:

Add digits

Write last digit

Carry forward

Same logic — just digit-by-digit in nodes.
'''

def addTwoNumbers(l1, l2):
    # Dummy node helps us build the result list easily
    dummy = ListNode(0)

    # 'curr' will always point to the last node of result list
    curr = dummy

    # carry stores the carry from previous addition
    carry = 0

    # Loop runs as long as there is something to add
    while l1 or l2 or carry:

        # If l1 exists, take its value; otherwise use 0
        v1 = l1.val if l1 else 0

        # If l2 exists, take its value; otherwise use 0
        v2 = l2.val if l2 else 0

        # Add both digits and carry
        total = v1 + v2 + carry

        # Update carry for next iteration
        carry = total // 10

        # Create a new node with the last digit of total
        curr.next = ListNode(total % 10)

        # Move curr forward
        curr = curr.next

        # Move l1 and l2 forward if possible
        if l1:
            l1 = l1.next
        if l2:
            l2 = l2.next

    # dummy.next is the real head of the result list
    return dummy.next
#--------------------------------------------------------------------------
class ListNode:
    def __init__(self, val):
        self.val = val
        self.next = None
def create_linked_list(arr):
    dummy = ListNode(0)
    curr = dummy

    for num in arr:
        curr.next = ListNode(num)
        curr = curr.next

    return dummy.next
def print_linked_list(head):
    curr = head
    while curr:
        print(curr.val, end=" → ")
        curr = curr.next
    print("None")

# Input numbers (reverse order)
l1 = create_linked_list([2, 4, 3])  # represents 342
l2 = create_linked_list([5, 6, 4])  # represents 465

# Add the two numbers
result = addTwoNumbers(l1, l2)

# Print result
print_linked_list(result)

l1 = create_linked_list([9, 9])   # 99
l2 = create_linked_list([1])      # 1

result = addTwoNumbers(l1, l2)
print_linked_list(result)
```

# 6. Intersection of Two Linked Lists — LeetCode 160

❓ **What is the question asking?**

Given two singly linked lists, find the node where they intersect.

🔍 **Key clarifications**

- Intersection means same node in memory, not same value

- If they do not intersect, return None

- Lists may have different lengths

📌 **Example**

image.png

🧠 **DSA Pattern**

Pointer Switching / Length Alignment

💡 **Approach**

Traverse both lists; switch heads when reaching end.

**Pseudocode**

image.png

```python
def getIntersectionNode(headA, headB):
    # Initialize two pointers
    # 'a' starts at the head of List A
    # 'b' starts at the head of List B
    a, b = headA, headB

    # Loop until both pointers point to the same node
    # This node could be the intersection node or None
    while a != b:

        # Move pointer 'a' to the next node
        # If 'a' reaches the end of List A (becomes None),
        # redirect it to the head of List B
        a = a.next if a else headB

        # Move pointer 'b' to the next node
        # If 'b' reaches the end of List B (becomes None),
        # redirect it to the head of List A
        b = b.next if b else headA

    # When the loop ends:
    # - either both pointers meet at the intersection node
    # - or both become None (no intersection)
    return a

#---------------------------------------------------------------------------------

class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
# Create input
# Common part (intersection)
intersect = ListNode(8)
intersect.next = ListNode(9)

# List A
headA = ListNode(1)
headA.next = ListNode(2)
headA.next.next = ListNode(3)
headA.next.next.next = intersect

# List B
headB = ListNode(4)
headB.next = ListNode(5)
headB.next.next = intersect

result = getIntersectionNode(headA, headB)
print(result.val)  # Output: 8

```

# 7. Palindrome Linked List — LeetCode 234
**What is the question asking?**

Check whether a linked list reads the same forward and backward.

🔍 **Key clarifications**

- You should aim for O(1) extra space

- Values must match in mirror positions

📌 **Example**

image.png

🧠 **DSA Pattern**

Fast–Slow Pointer + Reverse Second Half

💡 **Approach**

- Find middle

- Reverse second half

- Compare both halves

🧾 Pseudocode

image.png

```python
def isPalindrome(head):
  slow = fast = head

  while fast and fast.next:
      slow = slow.next
      fast = fast.next.next

  prev = None
  while slow:
      nxt = slow.next
      slow.next = prev
      prev = slow
      slow = nxt

  left, right = head, prev
  while right:
      if left.val != right.val:
          return False
      left = left.next
      right = right.next

  return True
```

```python
def isPalindrome(head):
  # Step 1: Find the middle of the linked list
  # slow moves 1 step, fast moves 2 steps
  slow = fast = head

  while fast and fast.next:
      slow = slow.next
      fast = fast.next.next

  # Step 2: Reverse the second half of the list
  prev = None
  while slow:
      nxt = slow.next      # store next node
      slow.next = prev     # reverse the link
      prev = slow          # move prev forward
      slow = nxt           # move slow forward

  # Step 3: Compare first half and reversed second half
  left, right = head, prev

  while right:  # only need to compare second half
      if left.val != right.val:
          return False     # not a palindrome
      left = left.next
      right = right.next

  # If all values matched
  return True

#----------------------------------------------------
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
head = ListNode(1)
head.next = ListNode(2)
head.next.next = ListNode(3)
head.next.next.next = ListNode(2)
head.next.next.next.next = ListNode(1)

print(isPalindrome(head))  # Output: True
```

# 8. Reverse Nodes in k-Group — LeetCode 25
❓ **What is the question asking?**

Reverse nodes of the linked list in groups of size k.

🔍 **Key clarifications**

- Only reverse complete groups

- If remaining nodes < k, leave them as-is

- You must not change node values

📌 **Example**

image.png

🧠 **DSA Pattern**

Recursive + Pointer Reversal

💡 **Approach**

Reverse first k nodes, then recurse on remaining list.

🧾 **Pseudocode**

image.png

```python
def reverseKGroup(head, k):
    curr = head
    count = 0

    while curr and count < k:
        curr = curr.next
        count += 1

    if count < k:
        return head

    prev = None
    curr = head

    for _ in range(k):
        nxt = curr.next
        curr.next = prev
        prev = curr
        curr = nxt

    head.next = reverseKGroup(curr, k)
    return prev
```

```python
def reverseKGroup(head, k):
    # Step 1: Check if there are at least k nodes
    curr = head
    count = 0

    while curr and count < k:
        curr = curr.next
        count += 1

    # If nodes are less than k, do not reverse
    if count < k:
        return head

    # Step 2: Reverse first k nodes
    prev = None
    curr = head

    for _ in range(k):
        nxt = curr.next    # store next node
        curr.next = prev   # reverse the link
        prev = curr        # move prev forward
        curr = nxt         # move curr forward

    # Step 3: Recursively reverse remaining list
    # head is now the last node of reversed group
    head.next = reverseKGroup(curr, k)

    # Step 4: prev is the new head of this group
    return prev

# ---------------------------------------------------------------------
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
head = ListNode(1)
head.next = ListNode(2)
head.next.next = ListNode(3)
head.next.next.next = ListNode(4)
head.next.next.next.next = ListNode(5)

k = 2
# Reverse in k groups
new_head = reverseKGroup(head, k)

def printList(head):
    curr = head
    while curr:
        print(curr.val, end=" -> ")
        curr = curr.next
    print("None")


# PRINT OUTPUT
printList(new_head)
```
